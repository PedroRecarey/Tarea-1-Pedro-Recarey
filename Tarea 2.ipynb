Escriba un código, sin usar ciclos for ni while, que calcule los cuadrados de los números enteros enttre 1 y 100.

```python
import numpy as np #importamos numpy para poder trabajar con los array 
matriz=np.linspace(1,100,100, dtype=int) #se crea una variable y ocupo linspace para poder generar una matriz con que empieze en 1 y termine en 100, ademas
                                  # que el espacio entre los numeros sea de uno en uno. Tambien colocamos dtype para poder definir que los datos sean int

matriz_resultante= matriz*matriz # generamos la multiplicacion de la matriz por si misma para obtener la matriz elevada al cuadrado 
print('la matriz es:', matriz)
print()
print('La matriz resultante sera:' ) # Finalmente colocamos este print, para que el resultado quede mas prolijo
print(matriz_resultante)
```
