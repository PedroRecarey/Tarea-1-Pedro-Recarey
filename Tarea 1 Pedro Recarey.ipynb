Ejercicio propuesto

```python

def fun(n):
    return 1/n

print(fun(0))
```

Desarrollo:

```python
def fun(n):  
    return 1/n

print(fun(1)) 
```

En este ejercicio, se puede apreciar que se crea una funcion llamada fun(n),luego se aplica el comando return 1/n , luego con print(fun()) le damos un valor al argumento para que se aplique en la expresion 1/n y arroje un resultado.
El problema en esta situacion es que se genera una division por cero en la cuarta linea y la manera de arreglarlo es cambiarle el argumento a print(fun()) a cualquier valor posible exeptuando el cero.
